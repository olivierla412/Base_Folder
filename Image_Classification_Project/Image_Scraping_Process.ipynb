```python
## Importing the libraries
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException
from selenium.webdriver.common.keys import Keys
from tqdm import tqdm
import os
import io
import requests
from PIL import Image
```

```python
driver = webdriver.Chrome('/Users/oliverla412/Desktop/chromedriver')
time.sleep(3)
driver.maximize_window()
```

## Sarees women

```python
# Opening the home page of flipkart using the above url
driver.get('https://www.flipkart.com/')
time.sleep(2)

#Asking the user to input the keywords to search.
user_input = input('Enter here for the products: ')
time.sleep(2)
input_srch = driver.find_element_by_xpath('//input[@class="_3704LK"]') ## getting the input search
input_srch.clear() # clean the input search
time.sleep(2)
input_srch.send_keys(user_input) # getting the user input
time.sleep(2)
driver.find_element_by_xpath('//button[@class="L0Z3Pu"]').click() # search user input
time.sleep(2)
btn_urls = driver.find_elements_by_xpath('//nav[@class="yFHi8N"]//a') ## fetch the urls of each button
time.sleep(2)

btn_urls_list=  [x.get_attribute('href') for x in btn_urls]
btn_urls_list[0] # printing the first item
```

### let's create more urls of the button page using for loops

```python
Sarees_women_pages= [] ## creatinf an empty list

for x in tqdm(range(1,26)): ## loop for 25 buttons urls
    x =str(x)
    Sarees_women_pages.append(f"https://www.flipkart.com/search?q=Sarees+women&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={x}")
len(Sarees_women_pages)
```

```python
Sarees_women_pages[:3]
```

```python
Sarees_women_src = [] ## creating an empty list

for x in tqdm(Sarees_women_pages):
    driver.get(x)
    time.sleep(3)
    src = driver.find_elements_by_xpath('//img[@class="_2r_T1I"]') ## getting the image src urls
    time.sleep(2)
    try:
        for x in src:
            Sarees_women_src.append(x.get_attribute('src')) 
    except:
        Sarees_women_src.append('-')
```

```python
len(Sarees_women_src)
```

### Getting the images

```python
os.chdir('/Users/oliverla412/Desktop/Place_project/Image_Classification_Project/Images/sarees/')
baseDir=os.getcwd()

for i, url in enumerate(Sarees_women_src):
    file_name = f"{i:150}.jpg"    
    try:
        image_content = requests.get(url).content

    except Exception as e:
        print(f"ERROR - COULD NOT DOWNLOAD {url} - {e}")

    try:
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file).convert('RGB')
        
        file_path = os.path.join(baseDir, file_name)
        
        with open(file_path, 'wb') as f:
            image.save(f, "JPEG", quality=85)
        print(f"SAVED - {url} - AT: {file_path}")
    except Exception as e:
        print(f"ERROR - COULD NOT SAVE {url} - {e}")
```
