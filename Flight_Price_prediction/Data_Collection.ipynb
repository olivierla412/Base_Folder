## click on the 0 stop label and collect all the data related


```python
driver = webdriver.Chrome('/Users/oliverla412/Desktop/chromedriver')
time.sleep(3)
driver.maximize_window()


## fetching the website
url_ = 'https://www.yatra.com/'
driver.get(url=url_)
time.sleep(4)
## click on search flight text label
driver.find_element_by_xpath('//input[@id="BE_flight_flsearch_btn"]').click()
time.sleep(6)

## creating a list containing these defferent days.

item = ["Wed, 27 Oct","Thu, 28 Oct","Fri, 29 Oct","Sat, 30 Oct","Sun, 31 Oct"]

Date1 =[]
Airline1 =[]
Departure_time1 =[]
Duration1 = []
Stops1 = []
Price1 = []

for x in tqdm(item):
    el1 = driver.find_element_by_xpath('//*[contains(text(),"'+ x +'")]')
    time.sleep(10)
    Date1.append(el1.text)
    time.sleep(2)
    el1.click()
    ## click on the More Filters text link
    time.sleep(6)
    driver.find_element_by_xpath("//*[contains(text(),'More Filters')]").click()
    time.sleep(3)
    ## click on the 0 stop label
    driver.find_element_by_xpath('/html/body/section[2]/section/section[1]/div/div[2]/div/div/div[1]/div/label[1]').click()
    time.sleep(4)
    ## click on the Apply Filter text link
    driver.find_element_by_xpath('/html/body/section[2]/section/section[1]/div/div[1]/div/div[2]/div/div/div[2]/div[3]/input').click()
    time.sleep(8)
    #date = driver.find_element_by_xpath('//*[contains(text(),"'+ x +'")]')
    

    airline = driver.find_elements_by_xpath('//span[@class="i-b text ellipsis"]')
 
    for x in airline:
        Airline1.append(x.text)

    dep_time = driver.find_elements_by_xpath('//div[@class="i-b pr"]')
    for i in dep_time:
        Departure_time1.append(i.text)

    duration = driver.find_elements_by_xpath('//p[@class="fs-12 bold du mb-2"]')
    for i in duration:
        Duration1.append(i.text)
    
    stop = driver.find_elements_by_xpath('//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]//span')
    for oli in stop:
        Stops1.append(oli.text)
    price = driver.find_elements_by_xpath('//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')
    for oli in price:
        Price1.append(oli.text)   
    time.sleep(3)    

driver.close()

data1 = pd.DataFrame({})
data1['Date1'] = Date1
data = pd.DataFrame({})
data['Airline1'] = Airline1
data['Departure_time1'] = Departure_time1
data['Duration1'] = Duration1
data['Stops1'] = Stops1
data['Price1'] = Price1
print(data1)
print(data)
```
